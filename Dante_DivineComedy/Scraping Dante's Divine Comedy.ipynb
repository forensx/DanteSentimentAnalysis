```python
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from requests import get
url = 'https://www.gutenberg.org/files/8800/8800-h/8800-h.htm'
response = get(url)
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)
```

## Scraping Paradiso

1. Find all tables
2. Select second table (3rd for Paradiso)
3. Find all a tags in table

```python
all_section_tables = html_soup.find_all('table')
```

```python
# Build array of a tags
paradiso = all_section_tables[2].find_all('a', href=True)
len(paradiso)
```

```python
print(paradiso[0:3])
```

```python
df = pd.DataFrame()
df['Link'] = [entry['href'] for entry in paradiso]
df['Canto #'] = [entry.text for entry in paradiso]
df.to_csv("Paradiso_Directory_Gutenberg.csv", index = False)
df.head() # just for ease of use later, if needed
```
